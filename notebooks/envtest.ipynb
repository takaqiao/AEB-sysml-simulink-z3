```python
# 导入标准库
import sys
import pandas as pd
import matplotlib # 导入主库以获取版本号
import matplotlib.pyplot as plt

# 打印当前Python解释器的路径，确认它在'sysml_env'中
print(f"Python 解释器路径: {sys.executable}")
print("-" * 50)

# 打印关键库的版本
print(f"Pandas 版本: {pd.__version__}")
# *** 此处为修正部分 ***
print(f"Matplotlib 版本: {matplotlib.__version__}")

# 创建一个简单的图表示例
print("\n正在测试 Matplotlib 绘图...")
plt.figure(figsize=(5, 2))
plt.plot([0, 1, 2], [2, 3, 1])
plt.title("Matplotlib Test Plot")
plt.show()
print("Matplotlib 绘图成功！")
```
